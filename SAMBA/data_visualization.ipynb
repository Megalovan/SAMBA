```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_init = np.load('w_init.npy', allow_pickle=True)
b_init = np.load('b_init.npy', allow_pickle=True)

w_finetuned = np.load('w_finetuned.npy', allow_pickle=True)
b_finetuned = np.load('b_finetuned.npy', allow_pickle=True)
```

```python
fig1 = plt.figure(num = 1, figsize = (6,4.5))
ax1 = fig1.add_subplot(311)
ax1.plot(b_init[0], '-.', label = 'b1_init')
ax1.plot(b_finetuned[0], '-.', label = 'b1_finetuned')
ax1 = fig1.add_subplot(312)
ax1.plot(b_init[1], '-.', label = 'b2_init')
ax1.plot(b_finetuned[1], '-.', label = 'b2_finetuned')
ax1 = fig1.add_subplot(313)
ax1.plot(b_init[2], '-.', label = 'b3_init')
ax1.plot(b_finetuned[2], '-.', label = 'b3_finetuned')
plt.show()
```

```python
fig2 = plt.figure(num = 1, figsize = (6,4.5))
ax2 = fig2.add_subplot(311)
ax2.plot(w_init[0], '-.', label = 'w1_init')
ax2.plot(w_finetuned[0], '-.', label = 'w1_finetuned')
ax2 = fig2.add_subplot(312)
ax2.plot(w_init[1], '-.', label = 'w2_init')
ax2.plot(w_finetuned[1], '-.', label = 'w2_finetuned')
ax2 = fig2.add_subplot(313)
ax2.plot(w_init[2], '-.', label = 'w3_init')
ax2.plot(w_finetuned[2], '-.', label = 'w3_finetuned')
plt.show()
```
